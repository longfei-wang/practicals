```python
import pandas as pd
users = pd.read_csv('profiles.csv')
```

```python
ucountries = list()

for i in users.itertuples():
    c = i[4]
    if c not in ucountries:
        ucountries.append(c)
```

```python
ages = list()

for i in users.itertuples():
    c = i[3]
    if (not np.isnan(c)):
        ages.append(float(c))
```

```python
ma = np.mean(ages)
```

```python
user_matrix = dict()

for i in users.itertuples():
    row = []
    
    id = i[1]
    if i[2] == 'm':
        g = 0
    elif i[2] == 'f':
        g = 1
    else:
        g = 0.5
    
    if not np.isnan(i[3]):
        a = i[3]
    else:
        a = ma
    
    c = i[4]
    
    row += [g]
    row += [a]
    
    tmp =[]
    try:
        for i in ucountries:
            tmp += [(c==i)*1]
    except:
        tmp = [0] * len(ucountries)
    
    row += tmp
    
    user_matrix[id]= row
```

```python
######################################################################################
import csv
train_file = 'train.csv'
train_data = {}

with open(train_file, 'r') as train_fh:
    train_csv = csv.reader(train_fh, delimiter=',', quotechar='"')
    next(train_csv, None)
    for row in train_csv:
        user   = row[0]
        artist = row[1]
        plays  = int(row[2])
        
        if not user in train_data:
            train_data[user] = {}
        
        train_data[user][artist] = plays
```

```python
def predict(user,artist,nu,na):
    s_artists = neigh.kneighbors(np.array(artist_matrix[artist]),na,return_distance=True)
    s_users = uneigh.kneighbors(np.array(user_matrix[user]),nu,return_distance=True)
    
    uuu = [(uids[i],idist) for i,idist in zip(s_users[1][0],s_users[0][0])]
    aaa = [(ids[j],jdist) for j,jdist in zip(s_artists[1][0],s_artists[0][0])]
    
    return uuu,aaa
```

```python
def get_score(quser,qartist,nu=200,na=10):


    result = predict(quser,qartist,nu,na)

    score = list()
    n = 0 
    for j,jdist in result[1]:
        for i,idist in result[0]:
            if not (i==quser and j==qartist):
                try:
                    score.append(train_data[i][j])
                    #print train_data[i][j],idist, jdist
                    n += 1
                except:
                    pass
                
                if n > 5:
                    return score
    
    if len(score) != 0:
        return score
    elif nu < 1000:
        return get_score(quser,qartist,nu*2,na*2)
    else:
        return user_medians[quser]
```
